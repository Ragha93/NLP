```python
import numpy as np                                  
import pandas as pd

#NLTK toolkits libraries
import nltk,string,re 
from nltk.corpus import wordnet
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords                   
from nltk.stem import PorterStemmer,LancasterStemmer,WordNetLemmatizer
from nltk import word_tokenize,sent_tokenize

#sklearn toolkits
from sklearn.feature_extraction.text import CountVectorizer          
from sklearn.feature_extraction.text import TfidfVectorizer    
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.naive_bayes import MultinomialNB

#tensorflow libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
data = pd.read_csv("Reviews.csv")
data = data.head(1000)
```

# NLTK Word tokenizer

```python
[word_tokenize(i) for i in data.Text]
```

```python
[sent_tokenize(i) for i in data.Text]
```

# Removing the stopwords, Punctuation and  normalizing the sentences

```python
for i in data.Text:
    i = i.lower()
    i = re.sub("<.*>|[\s]{2,}| [.0-9]+| [.0-9]{2,10}|[^a-zA-Z]"," ",i)
    i = [t for t in word_tokenize(i) if t not in string.punctuation and t not in stopwords.words('english')]
    i = " ".join(i)
#     i = [t for t in i if t not in string.punctuation]
```

```python
b = [" ".join([t for t in word_tokenize(re.sub("<.*>|[\s]{2,}| [.0-9]+| [.0-9]{2,10}|[^a-zA-Z]"," ",i.lower())) if t not in string.punctuation and t not in stopwords.words('english')]) for i in data.Text]
```

```python
len(b)
```

# Stemming and Lemmatizing

```python
X = b
y = data.Score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
PS = PorterStemmer()
LS = LancasterStemmer()
WL = WordNetLemmatizer()
```

```python
X_train_s = []
for i in X_train:
    m = []
    for j in word_tokenize(i):
        m.append(PS.stem(j))
    X_train_s.append(' '.join(m))

X_test_s = []
for i in X_test:
    m = []
    for j in word_tokenize(i):
        m.append(PS.stem(j))
    X_test_s.append(' '.join(m))
    
#Porter Stemmer    
```

```python
X_train_ls = []
for i in X_train:
    m = []
    for j in word_tokenize(i):
        m.append(LS.stem(j))
    X_train_ls.append(' '.join(m))

X_test_ls = []
for i in X_test:
    m = []
    for j in word_tokenize(i):
        m.append(LS.stem(j))
    X_test_ls.append(' '.join(m))
    
#Lancaster Stemmer    
```

```python
X_train_l = []
for i in X_train:
    m = []
    for j in word_tokenize(i):
        m.append(WL.lemmatize(j))
    X_train_l.append(' '.join(m))

X_test_l = []
for i in X_test:
    m = []
    for j in word_tokenize(i):
        m.append(WL.lemmatize(j))
    X_test_l.append(' '.join(m))
    
#Wordnet lemmatizer    
```

```python
CV = CountVectorizer(tokenizer=word_tokenize,stop_words=stopwords.words('english'))
TF = TfidfVectorizer()
```

# Simple Multinomial Naive bayes using Count vectorizer on stemmed data

```python
X_train_1 = CV.fit_transform(X_train_s)
X_test_1 = CV.transform(X_test_s)

X_train_1t = TF.fit_transform(X_train_s)
X_test_1t = TF.transform(X_test_s)
```

```python
MNB = MultinomialNB()
MNB.fit(X_train_1,y_train)
pred = MNB.predict(X_test_1)
print(classification_report(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```
