```python
text = """In the year 1960, APJ Abdul Kalam’s graduation took place from Madras Institute of Technology. The association of Kalam took place with the Defence Research & Development Service (DRDS). Furthermore, he joined as a scientist at the Aeronautical Development Establishment of the Defence Research and Development Organisation. These were the beginning achievements of his prestigious career as a scientist.
Big achievement for Kalam came when he was the project director at ISRO of India‘s first-ever Satellite Launch Vehicle (SLV- III). This satellite was responsible for the deployment of the Rohini satellite in 1980. Moreover, Kalam was highly influential in the development of Polar Satellite Launch Vehicle (PSLV) and SLV projects.
Both projects were successful. Bringing enhancement in the reputation of Kalam. Furthermore, the development of ballistic missiles was possible because of the efforts of this man. Most noteworthy, Kalam earned the esteemed title of “The missile Man of India”. Kalam KalamKalam
email: raga1193@gmail.com and 919962207216 and 9962207216 <body> This is the great Kalam </body>"""
```

# Simple Regular Expressions

# 1. Finding the former president's name word Kalam in paragraph

```python
pattern = re.compile(r'Kalam')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 2. Finding all characters including space except new line

```python
pattern = re.compile(r'.')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 3. Identifying fullstop.  Anything that comes after \ shall be found

```python
pattern = re.compile(r'\.')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 4. Identifying Digits using small d. Anything that comes after \ shall be found and here \d digits shall be found

```python
pattern = re.compile(r'\d')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 5. Identifying Non-Digits using Caps D. Anything that comes after \ shall be found and here \D non-digits shall be found which includes space and newline

```python
pattern = re.compile(r'\D')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 6. Find Digits with word characters, non-overlapping ones

## Eg: text="123456789a" 
## pattern = re.compile(r'\d\w')
## pattern = pattern.finditer(text)
## for p in pattern: print(p)
## This will only identify 12 34 56 78 9a  
### and not 12 23 34 45 etc

```python
pattern = re.compile(r'\d\w')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 7. \s identifies whitespace, tabspace, newline

```python
pattern = re.compile(r'\s')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 8. \S identifies non-whitespaces

```python
pattern = re.compile(r'\S')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# Word Boundary - Finding words that either have a boundary or nothing.

# 1. Finding the former president's name word kalam without boundary

```python
pattern = re.compile(r'Kalam')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 2. Finding the former president's name word kalam with a boundary
## If there is Kalam it will be detected but if it is KalamKalam then only one Kalam will be found as it has boundary whereas one of it is closely packed with no boundary

```python
pattern = re.compile(r"Kalam\b")
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 3. Finding the former president's name word kalam with a boundary on both sides.

## If there is Kalam with a boundary before and after it then it will be detected

```python
pattern = re.compile(r'\bKalam\b')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 4. Finding the former president's name word kalam with a non-boundary in left and word bounday on the right.

## If there is Kalam with a non-wordboundary before and a word boundary after it then it will be detected

## eg: In sentence "Kalam KalamKalam" the last word Kalam has no word boundary in the left as it is stuck to another word but ends with a word boundary, hence it will be considered

```python
pattern = re.compile(r'\BKalam\b')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 5. Finding the digit with a word boundary before it

```python
pattern = re.compile(r'\b\d')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 6. Anything that starts with Non-White space

```python
pattern = re.compile(r'^\S')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# Character sets- comes with square brackets [], anything inside square brackets will be found

# 1. Search for 1960 but has word boundary to its right

```python
pattern = re.compile(r'[1960]\w')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 2. Search for 2 characters that is between a to z

```python
pattern = re.compile(r'[a-z][a-z]')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 3. Searching for a string that starts with ( and some alphabets and then ends with ). 

## Eg: (DRDS), (SLV) and (PSLV)

```python
pattern = re.compile(r'\([A-Z)]+')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# 4. Searching for a phone number and email id

## Eg:

```python
pattern = re.compile(r'[a-z]+[0-9]+[@]+[a-z]+\.[a-z]+')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

```python
pattern = re.compile(r'\d{12}')
pattern = pattern.finditer(text)
for p in pattern:
    print(p)
```

# Accessing the information - While iterating get the .group(0) output

```python
pattern = re.compile(r'[a-z0-9]+@[a-z]+\.[a-z]+')
pattern = pattern.finditer(text)
for p in pattern:
    print(p.group(0))
```

```python
pattern = re.compile(r'</?[a-z]+>')
pattern = pattern.finditer(text)
for p in pattern:
    print(p.group(0))
```

## General

### 1) .  : Any Character except new line
### 2)\d : Digits (0-9)
### 3)\D : Non-digit
### 4)\w : Word Character (a-z,A-Z,0-9,_)
### 5)\W: NonWord Character
### 6)\s : Whitespace (Space, tab, newline)
### 7)\S: Non-Whitespace



## Word Boundary
### 1) \b : Word Boundary
### 2) \B: Non-Word Boundary
### 3) ^: Beginning of a string
### 4)  $ : End of a string


## Quantifiers:
### 1) *: 0 or more times
### 2) +: 1 or more times
### 3) ?: 0 or 1 times
### 4){3}: Exact number of times
### 5){3,4}: Range of numbers {min,max}
### 6)[]: Matches the characters in brackets
### 7)[^]: Matches characters not in brackets
### 8)| : Either or
### 9)(): Group of
